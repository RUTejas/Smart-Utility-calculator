```python
# Smart Utility Calculator - Basic Python Topics Only

def arithmetic_calculator():
    print("\n--- Arithmetic Calculator ---")
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    print("Addition: " + str(a + b))
    print("Subtraction: " + str(a - b))
    print("Multiplication: " + str(a * b))
    print("Division: " + str(a / b))

def bitwise_operations():
    print("\n--- Bitwise Operations ---")
    x = int(input("Enter first number: "))
    y = int(input("Enter second number: "))
    print("x & y = " + str(x & y))
    print("x | y = " + str(x | y))
    print("x ^ y = " + str(x ^ y))
    print("~x = " + str(~x))

def unit_converter():
    print("\n--- Unit Converter (cm to inches) ---")
    cm = float(input("Enter length in cm: "))
    inches = cm / 2.54
    print("Length in inches: " + str(round(inches, 2)))

def swap_variables():
    print("\n--- Swap Two Variables ---")
    a = input("Enter first value: ")
    b = input("Enter second value: ")
    print("Before swapping: a = " + a + ", b = " + b)
    a, b = b, a
    print("After swapping: a = " + a + ", b = " + b)

def round_value():
    print("\n--- Round Off Value ---")
    num = float(input("Enter a decimal number: "))
    print("Rounded value: " + str(round(num)))

def number_properties_checker():
    print("\n--- Number Properties Checker ---")
    n = int(input("Enter a number: "))
    if n % 2 == 0:
        print(str(n) + " is even")
    else:
        print(str(n) + " is odd")

    if n > 0:
        print(str(n) + " is positive")
    elif n < 0:
        print(str(n) + " is negative")
    else:
        print("It's zero")

def identity_checker():
    print("\n--- Identity Checker ---")
    x = input("Enter a value: ")
    y = input("Enter another value: ")
    if x is y:
        print("x is y")
    else:
        print("x is not y")

def membership_checker():
    print("\n--- Membership Checker ---")
    text = input("Enter a string: ")
    char = input("Enter character to check: ")
    if char in text:
        print(char + " is in the string.")
    else:
        print(char + " is not in the string.")

def string_concatenation():
    print("\n--- String Concatenation ---")
    first = input("Enter first word: ")
    second = input("Enter second word: ")
    message = "Combined string: " + first + " " + second
    print(message)

def greeting_generator():
    print("\n--- Custom Greeting ---")
    name = input("Enter your name: ")
    print("Hello " + name + "! Welcome to Smart Calculator.")

# Main menu
while True:
    print("\n--- Smart Utility Calculator ---")
    print("1. Arithmetic Calculator")
    print("2. Bitwise Operations")
    print("3. Unit Converter")
    print("4. Swap Two Variables")
    print("5. Round Off Value")
    print("6. Number Properties Checker")
    print("7. Identity Checker")
    print("8. Membership Checker")
    print("9. String Concatenation")
    print("10. Greeting Generator")
    print("11. Exit")

    choice = input("Enter your choice (1-11): ")

    if choice == "1":
        arithmetic_calculator()
    elif choice == "2":
        bitwise_operations()
    elif choice == "3":
        unit_converter()
    elif choice == "4":
        swap_variables()
    elif choice == "5":
        round_value()
    elif choice == "6":
        number_properties_checker()
    elif choice == "7":
        identity_checker()
    elif choice == "8":
        membership_checker()
    elif choice == "9":
        string_concatenation()
    elif choice == "10":
        greeting_generator()
    elif choice == "11":
        print("Thank you for using Smart Calculator. Goodbye!")
        break
    else:
        print("Invalid choice. Try again.")
```
